```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import warnings
warnings.filterwarnings("ignore")
```

# Read data

```python
actions = pd.read_csv("../data/user16_vs_the_world.csv")
actions.head()
```

```python
actions.columns
```

```python
# split into features and target variable
X = actions.drop("user_16", axis=1)
y = actions["user_16"]

norm_X = normalize(X)

X_train, X_test, y_train, y_test = train_test_split(norm_X, y, test_size=0.3,
                                                    random_state=42)
```

# Let's try a PCA

```python
pca = PCA(n_components=3)
```

```python
pca.fit(X_train)
```

```python
print(pca.explained_variance_ratio_)
```

```python
X_train_reduce = pca.fit_transform(X_train)
```

```python
knn_pca = KNeighborsClassifier(n_neighbors=33)
```

```python
knn_pca.fit(X_train_reduce, y_train)
```

```python
knn_pca.score(X_train_reduce, y_train)
```

```python
pca_results = pd.DataFrame(X_train_reduce, columns=["pca1", "pca2", "pca3"])
```

```python
temp_df_pca = actions.join(pca_results)
temp_df_pca.head()
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs=temp_df_pca["pca1"],
           ys=temp_df_pca["pca2"],
           zs=temp_df_pca["pca3"],
           c=temp_df_pca["user_16"])

ax.set_xlabel("PCA Component 1")
ax.set_ylabel("PCA Component 2")
ax.set_zlabel("PCA Component 3")

plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(24, 8))

sns.scatterplot(x=temp_df_pca["pca1"], y=temp_df_pca["pca2"], hue=temp_df_pca["user_16"], ax=ax1)
ax1.set_xlabel("PCA Component 1")
ax1.set_ylabel("PCA Component 2")

sns.scatterplot(x=temp_df_pca["pca1"], y=temp_df_pca["pca3"], hue=temp_df_pca["user_16"], ax=ax2)
ax2.set_xlabel("PCA Component 1")
ax2.set_ylabel("PCA Component 3")

sns.scatterplot(x=temp_df_pca["pca2"], y=temp_df_pca["pca3"], hue=temp_df_pca["user_16"], ax=ax3)
ax3.set_xlabel("PCA Component 2")
ax3.set_ylabel("PCA Component 3")

plt.show()
```

# K-Nearest Neighbors using Hamming distance

```python
scores = []

for i in range(1, 50, 2):
    knn_ham = KNeighborsClassifier(n_neighbors=i, metric="hamming")
    knn_ham.fit(X_train, y_train)
    scores.append(knn_ham.score(X_test, y_test))
```

```python
# plot scores for different numbers of neighbors

fig, ax = plt.subplots(figsize=(16, 8))
sns.lineplot(x=list(range(1, 50, 2)), y=scores, ax=ax, lw=3)
ax.set_title("Searching for the Plateau", color="black", size=20, y=1.03)
ax.set_xlabel("Number of Neighbors", color="black", size=20)
ax.set_ylabel("Score", color="black", size=20)
ax.tick_params(colors="black")

ax.annotate("Plateau here?",
            xy=(0.58, 0.85),
            xytext=(0.55, 0.4),
            xycoords="figure fraction",
            textcoords="figure fraction",
            size=20,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle='arc3',
                            color='grey',
                            lw=3))

plt.show()

# knn suggests 28 neighbors might be best
```
